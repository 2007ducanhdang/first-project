```python
#W6A16
n,m=map(int,input().split())
mat=""
for i in range(n):
    for j in range(m):
        item=int(input())
        mat+=(f'{item:>4}')
    mat+='\n'
print(mat)
        
```

```python
#W6A17
n=int(input())
mat=[]
for i in range(n):
    hang=[]
    for j in range(n):
        item=int(input())
        hang.append(item)
    mat.append(hang)
cheochinh=[mat[i][i] for i in range(len(mat))]
cheophu=[mat[j][len(mat)-1-j] for j in range(len(mat))]
print(mat)
print(cheochinh)
print(cheophu)
```

```python
#W6A18
m,n,k=map(int,input().split())
mat=[]
for i in range(n):
    hang=[]
    for j in range(m):
        item=int(input())
        hang.append(item)
    mat.append(hang)
tonghangk=sum(mat[k-1])
tongcotk=sum([mat[i][k-1] for i in range(len(mat))])
print(mat,tonghangk,tongcotk)
```

```python
#W6A21
A=input().split()
dic={}
for i in range(len(A)):
    if A[i] not in dic:
        dic[A[i]]=[i]
    else:
        dic[A[i]].append(i)
print(dic)
```

```python
#W6A23
def xoa_so(A,x):
    B=[i for i in A if i>=x]
    return B
    
A=list(map(int,input().split()))
x=int(input())

print(xoa_so(A,x))
```

```python
#W6A24
def day_so(A):
    A0=[]
    for i in A:
        if i==0:
            A0.append(i)
    B=[j for j in A if j!=0]
    return A0+B
A=list(map(int,input().split()))
print(day_so(A))
```
