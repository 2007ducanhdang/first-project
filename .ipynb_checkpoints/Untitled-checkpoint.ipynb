```python
#W5A1
a,b=map(int,input().split())
def Max_of_two(a,b):
    return max(a,b)
print(Max_of_two(a,b))
```

```python
#W5A2
a,b=map(int,input().split())
def swap(a,b):
    return (b,a)
print(swap(a,b))
```

```python
#W5A3
a=int(input())
def songuyento(a):
    ans=False
    if a<2:
        ans=False
    elif a==2:
        ans=True
    else:
        dem=0
        for i in range(2,int(a**0.5)):
            if a%i==0:
                dem+=1
        if dem>0:
            ans=False
        else:
            ans=True
    return ans
print(songuyento(a))
```

```python
#W5A4
a=int(input())
def sohoanhao(a):
    tguc=0
    for i in range(1,a):
        if a%i == 0:
            tguc+=i
    if tguc==a:
        return True
    else:
        return False
print(sohoanhao(a))      
    
```

```python
#W5A5
a=int(input())
def timkiem(k):
    A=list(map(int,input().split()))
    if k in A:
        return A.index(k)
    else:
        return -1
print(timkiem(a))
```

```python
#W5A6
b=int(input())
def giaithua(n):
    a=1
    for i in range(1,n+1):
       a=a*i
    return a
print(giaithua(b))    
    
        
    
```

```python
#W5A7
def maytinh(a,b):
    a,b=map(float,input().split())
    tong = round(a + b, 2)
    tru = round(a - b, 2)
    nhan = round(a*b, 2)
    chia = round(a/b, 2)
    c=input()
    if c == '+':
        print(tong)
    if c== '-':
        print(tru)
    if c== '*':
        print(nhan)
    if c== '/':
        print(chia)
maytinh(a,b)    


```

```python
def tong_chu_so(n):
    A=0
    while n>0:
        i=n%10
        A+=i
        n//=10
    return A
a=int(input())
print(tong_chu_so(a))
        
    
```
