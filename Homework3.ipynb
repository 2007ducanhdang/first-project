```python
#bai4
h=abs(int(input()))
for i in range(h+1):
    l=h-i-1
    r=h+i-1
    for j in range(2*h-1):
        if j==l or j==r or i==h-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
        
```
