```python
#W9A1
A=list(map(int,input().split()))
B=list(map(int,input().split()))
def merge_sort(a,b):
    res=[]
    i=j=0
    while i<len(a) and j<len(b):
        if a[i]<b[j]:
            res.append(a[i])
            i+=1
        else:
            res.append(b[j])
            j+=1
    res.extend(A[i:])
    res.extend(B[j:])
    return res
print(merge_sort(A,B))
        
```

```python
#W9A3
def doicho(A):
    a=list(A)
    for i in range(len(a)):
        min_index=i
        for j in range(i+1,len(a)):
            if a[j]<=a[min_index]:
                min_index=j
        if a[min_index]<a[i]:
            a[i],a[min_index]=a[min_index],a[i]
            break
    return "".join(a)
a=input()
n=list(a)
print(doicho(a))
            
            
```

```python
#W9A4
a = input()
b = input()
n = len(a)
m = len(b)
a = " " + a
b = " " + b
dp = [[0] * (m + 1) for _ in range(n + 1)]
dp[0][0] = 1
for i in range(1, n + 1):
    for j in range(m + 1):
        if dp[i - 1][j] and ('a' <= a[i] <= 'z'):
            dp[i][j] = 1
        if j > 0 and dp[i - 1][j - 1]:
            if a[i] == b[j] or a[i].upper() == b[j]:
                dp[i][j] = 1
if dp[n][m]:
    print("YES")
else:
    print("NO")
```

```python
#W9A6
n=list(map(int,input().split()))
reverse_n=n[::-1]
key=-1
for i in range(len(reverse_n)-1):
    if reverse_n[i]>reverse_n[i+1]:
        key=i+1
        break
if key==-1:
    print(" ".join(map(str,reverse_n)))
else:
    pivot=len(reverse_n)-key-1
    swap=pivot
    for j in range(swap+1,len(n)):
        if n[j]>n[pivot]:
            swap=j
    n[pivot],n[swap]=n[swap],n[pivot]
    ans=n[:pivot+1]+n[pivot+1:][::-1]
    kq=" ".join(map(str,ans))
    print(kq)
    

```
