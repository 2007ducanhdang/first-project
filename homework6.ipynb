```python
#W6A4
A = input("Nhập chuỗi: ").replace(":"," ").split()
dic = {}
for i in range(0, len(A), 2):
    key = A[i]
    val = A[i+1]
    
    if key not in dic:
        dic[key] = [val]
    else:
        dic[key].append(val)

print(dic)
```

```python
#W6A10
A=list(map(int,input().split()))
k=int(input())
ans=[]
for i in range(len(A)-1):
    for j in range(i+1,len(A)):
        if A[i]+A[j]==k and (A[i],A[j]) not in ans:
            ans.append((A[i],A[j]))
ans.sort()
print(ans)
```
