```python
#W7A1
def binary_search(arr,target):
    left=0
    right=len(arr)-1
    while left<=right:
        mid=(left+right)//2
        if target==arr[mid]:
            return mid
        elif target>arr[mid]:
            left=mid+1
        else:
            right=mid-1
    return -1
A=list(map(int,input().split()))
n=int(input())
print(binary_search(A,n))
```

```python
#W7A3
def bubblesort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
def selectionsort(arr):
    for i in range(len(arr)):
        moc=i
        for j in range(i+1,len(arr)):
            if arr[j]<arr[moc]:
                moc=j
        arr[i],arr[moc]=arr[moc],arr[i]
    return arr
A=list(map(int,input().split()))
print(selectionsort(A))
            
```

```python
#W7A4
def countmax(arr):
    a=set(arr)
    ans=0
    dem=0
    for i in a:
        if arr.count(i)>dem:
            ans=i
            dem=arr.count(i)
    return ans, arr.count(ans)
A=list(map(int,input().split()))
print(countmax(A))
```

```python
#W7A5
A=list(map(int,input().split()))
n=int(input())
dic={}
ans=0
for i in A:
    if i in dic:
        dic[i]+=1
    else:
        dic[i]=1
for j in dic.keys():
    if n-j in dic:
        if n-j==j:
            ans+=(dic[j]*(dic[j]-1))
        else:
            ans+=(dic[j]*dic[n-j])
print(ans//2)
```

```python
#W7A7
A=list(map(int,input().split()))
L=[1]*len(A)
for i in range(len(A)):
    for j in range(i):
        if A[i]>A[j]:
            L[i]=max(L[i],L[j]+1)
print(max(L))
```

```python
#W8A8
import ast
A=ast.literal_eval(input())
B=ast.literal_eval(input())
def binarysearch(arr,n):
    left=0
    right=len(arr)-1
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==n:
            return mid
        elif arr[mid]<n:
            left=mid+1
        else:
            right=mid-1
    return -1
ans=[9999]*len(B)
for i in A:
    x=[]
    for j in range(i[0],i[1]+1):
        x.append(binarysearch(B,j))
    ans[j]=min(ans[j],len(x))
print(ans)
```
