```python
#W8A2
import sys

class Date:
    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y

    def is_valid(self):
        if self.m < 1 or self.m > 12: return False
        days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.is_leap(self.y): days[2] = 29
        if self.d < 1 or self.d > days[self.m]: return False
        return True

    def is_leap(self, y):
        return (y % 400 == 0) or (y % 4 == 0 and y % 100 != 0)

    def next_day(self):
        if not self.is_valid(): return "INVALID"
        
        days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.is_leap(self.y): days[2] = 29
        
        self.d += 1
        if self.d > days[self.m]:
            self.d = 1
            self.m += 1
            if self.m > 12:
                self.m = 1
                self.y += 1
        
        return f"{self.d:02d}/{self.m:02d}/{self.y}"

def solve():
    input_data = sys.stdin.read().splitlines()
    for line in input_data:
        if not line.strip(): continue
        try:
            d, m, y = map(int, line.split('/'))
            date_obj = Date(d, m, y)
            if not date_obj.is_valid():
                print("INVALID")
            else:
                print(date_obj.next_day())
        except ValueError:
            print("INVALID")

if __name__ == "__main__":
    solve()
```

```python
#W8A3
import math
import sys

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def check_position(self):
        is_origin = (self.x == 0 and self.y == 0)
        on_x = (self.y == 0)
        on_y = (self.x == 0)
        return is_origin, on_x, on_y

    def distance_to_origin(self):
        return math.sqrt(self.x**2 + self.y**2)

def solve():
    try:
        line = sys.stdin.read().strip()
        if not line: return
        x, y = map(float, line.split())
        p = Point(x, y)
        
        is_origin, on_x, on_y = p.check_position()
        dist = p.distance_to_origin()

        if is_origin:
            print("La goc toa do")
        else:
            if on_x: print("Nam tren truc hoanh")
            elif on_y: print("Nam tren truc tung")
            else: print("Khong nam tren truc toa do")
            
        print(f"{dist:.2f}")

    except ValueError:
        pass

if __name__ == "__main__":
    solve()
```

```python
#W8A4
import sys

class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def set_numbers(self, a, b):
        self.a = a
        self.b = b

    def add(self): return self.a + self.b
    def subtract(self): return self.a - self.b
    def product(self): return self.a * self.b
    def divide(self): return self.a / self.b if self.b != 0 else "Div0"
    def power(self): return self.a ** self.b
    def mod(self): return self.a % self.b if self.b != 0 else "Div0"

def solve():
    lines = sys.stdin.read().splitlines()
    if not lines: return
    
    try:
        a, b = map(float, lines[0].split())
        calc = Calculator(a, b)
        
        i = 1
        while i < len(lines):
            cmd = lines[i].strip()
            if cmd.lower() in ["yes"]:
                break
            if cmd == '1': print(calc.add())
            elif cmd == '2': print(calc.subtract())
            elif cmd == '3': print(calc.product())
            elif cmd == '4': print(calc.divide())
            elif cmd == '5': print(calc.power())
            elif cmd == '6': print(calc.mod())
            elif cmd.startswith('7'):
                parts = cmd.split()
                if len(parts) == 3:
                    calc.set_numbers(float(parts[1]), float(parts[2]))
            i += 1
    except:
        pass

if __name__ == "__main__":
    solve()
```

```python
#W8A6
import sys

class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    
    def area(self):
        return self.w * self.h
    
    def perimeter(self):
        return 2 * (self.w + self.h)
    
    def scale(self, k):
        self.w *= k
        self.h *= k

def solve():
    input_data = sys.stdin.read().split()
    if not input_data: return
    
    try:
        w = float(input_data[0])
        h = float(input_data[1])
        k = float(input_data[2])
        
        rect = Rectangle(w, h)
        rect.scale(k)
        
        print(f"{rect.area():.2f} {rect.perimeter():.2f}")
    except:
        pass

if __name__ == "__main__":
    solve()
```

```python
#W8A7
import sys
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0: raise ValueError
        g = math.gcd(num, den)
        self.num = num // g
        self.den = den // g
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def __str__(self):
        return f"{self.num}/{self.den}"

    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den, self.den * other.den)

    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den, self.den * other.den)

    def mul(self, other):
        return Fraction(self.num * other.num, self.den * other.den)

    def div(self, other):
        return Fraction(self.num * other.den, self.den * other.num)

def solve():
    lines = sys.stdin.read().splitlines()
    for line in lines:
        if not line.strip(): continue
        parts = line.split()
        if len(parts) != 4: continue
        
        try:
            a, b = int(parts[0]), int(parts[1])
            op = parts[2]
            c, d = int(parts[3]), int(parts[4]) if len(parts) > 4 else int(parts[3]) # Logic parsing check

            c = int(parts[3]) 
            
            f1 = Fraction(a, b)

            
            tokens = line.split()
            a = int(tokens[0])
            b = int(tokens[1])
            op = tokens[2]
            c = int(tokens[3])
            d = int(tokens[4])
            
            f1 = Fraction(a, b)
            f2 = Fraction(c, d)
            
            if op == '+': print(f1.add(f2))
            elif op == '-': print(f1.sub(f2))
            elif op == '*': print(f1.mul(f2))
            elif op == '/': print(f1.div(f2))
            
        except:
            pass

if __name__ == "__main__":
    solve()
```

```python
#W8A8
import sys

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount

    def withdraw(self, amount):
        if amount > 0 and self.balance >= amount:
            self.balance -= amount

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

def solve():
    lines = sys.stdin.read().splitlines()
    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if not line: 
            i += 1
            continue
        
        parts = line.split()
        owner = parts[0]
        balance = float(parts[1])
        account = BankAccount(owner, balance)
        
        i += 1
        if i >= len(lines): break
        try:
            n = int(lines[i].strip())
            i += 1
            for _ in range(n):
                if i < len(lines):
                    cmd = lines[i].split()
                    op = cmd[0]
                    val = float(cmd[1])
                    if op == 'DEPOSIT': account.deposit(val)
                    elif op == 'WITHDRAW': account.withdraw(val)
                    i += 1
            print(f"{account.balance:.2f}")
        except:
            break

if __name__ == "__main__":
    solve()
```

```python
#W8A9
import sys

class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(score)

    def avg(self):
        if not self.scores: return 0.0
        return sum(self.scores) / len(self.scores)

    def rank(self):
        a = self.avg()
        if a >= 8.0: return "Excellent"
        if a >= 6.5: return "Good"
        if a >= 5.0: return "Average"
        return "Poor"

def solve():
    lines = sys.stdin.read().splitlines()
    i = 0
    while i < len(lines):
        name = lines[i].strip()
        if not name: 
            i += 1
            continue
            
        student = Student(name)
        i += 1
        if i >= len(lines): break
        try:
            n = int(lines[i].strip())
            i += 1
            for _ in range(n):
                if i < len(lines):
                    parts = lines[i].split()
                    score = float(parts[-1])
                    student.add_score("", score)
                    i += 1
            print(f"{student.name} {student.avg():.2f} {student.rank()}")
        except:
            break

if __name__ == "__main__":
    solve()
```

```python
#W8A10
import sys

class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, author):
        return sum(1 for b in self.books if b.author == author)

    def find_by_year(self, year):
        return sum(1 for b in self.books if b.year == year)

def solve():
    lines = sys.stdin.read().splitlines()
    lines = [l.strip() for l in lines if l.strip()]
    if not lines: return
    
    n = int(lines[0])
    lib = Library()
    
    for k in range(1, len(lines)):
        cmd = lines[k]
        if cmd.startswith("ADD"):
            # Format: ADD title; author; year
            parts = cmd[4:].split(';')
            if len(parts) == 3:
                lib.add(Book(parts[0].strip(), parts[1].strip(), int(parts[2].strip())))
        elif cmd.startswith("COUNTYEAR"):
            y = int(cmd.split()[1])
            print(lib.find_by_year(y))
        elif cmd.startswith("COUNT"):
            # COUNT author
            auth = cmd[6:].strip()
            print(lib.count_by_author(auth))

if __name__ == "__main__":
    solve()
```

```python
#W8A11
import sys
import math

class Vector3D:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z
    
    def sub(self, other):
        return Vector3D(self.x - other.x, self.y - other.y, self.z - other.z)
        
    def cross(self, other):
        return Vector3D(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x
        )
    
    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
        
    def mag(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

def solve():
    input_data = sys.stdin.read().split()
    if not input_data: return
    coords = [float(x) for x in input_data]
    
    # 4 points, 3 coords each -> 12 numbers per test case
    for i in range(0, len(coords), 12):
        chunk = coords[i:i+12]
        if len(chunk) < 12: break
        
        A = Vector3D(chunk[0], chunk[1], chunk[2])
        B = Vector3D(chunk[3], chunk[4], chunk[5])
        C = Vector3D(chunk[6], chunk[7], chunk[8])
        D = Vector3D(chunk[9], chunk[10], chunk[11])
        
        # Plane ABC normal: n1 = AB x AC
        AB = B.sub(A)
        AC = C.sub(A)
        n1 = AB.cross(AC)
        
        # Plane BCD normal: n2 = BC x BD
        BC = C.sub(B)
        BD = D.sub(B)
        n2 = BC.cross(BD)
        
        dot_val = abs(n1.dot(n2))
        mag_val = n1.mag() * n2.mag()
        
        if mag_val == 0:
            print("0.00")
        else:
            cos_theta = dot_val / mag_val
            # Clamp for safety
            if cos_theta > 1.0: cos_theta = 1.0
            angle_rad = math.acos(cos_theta)
            angle_deg = math.degrees(angle_rad)
            print(f"{angle_deg:.2f}")

if __name__ == "__main__":
    solve()
```

```python
#W8A12
import sys

class ComplexNumber:
    def __init__(self, r, i):
        self.r = r
        self.i = i
        
    def multiply(self, other):
        # (a+bi)(c+di) = (ac-bd) + (ad+bc)i
        real = self.r * other.r - self.i * other.i
        imag = self.r * other.i + self.i * other.r
        return ComplexNumber(real, imag)

def solve():
    data = sys.stdin.read().split()
    if not data: return
    
    try:
        r1, i1 = float(data[0]), float(data[1])
        r2, i2 = float(data[2]), float(data[3])
        
        c1 = ComplexNumber(r1, i1)
        c2 = ComplexNumber(r2, i2)
        res = c1.multiply(c2)

        print(f"{int(res.r) if res.r.is_integer() else res.r} {int(res.i) if res.i.is_integer() else res.i}")
        
    except:
        pass

if __name__ == "__main__":
    solve()
```

```python
#W8A13
import sys
import math

class Triangle:
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3
        
    def getArea(self):
        a, b, c = self.e1, self.e2, self.e3
        if a + b > c and a + c > b and b + c > a:
            p = (a + b + c) / 2
            return math.sqrt(p * (p - a) * (p - b) * (p - c))
        return -1

def solve():
    data = sys.stdin.read().split()
    if not data: return
    
    try:
        a, b, c = int(data[0]), int(data[1]), int(data[2])
        t = Triangle(a, b, c)
        area = t.getArea()
        
        if area == -1:
            print("invalid")
        else:
            print(f"{area:.2f}")
    except:
        pass

if __name__ == "__main__":
    solve()
```
