```python
#W8A1
class Cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def S(self):
        pi=3.14
        return f'{2*pi*self.r*(self.r+self.h):.2f}'
    def V(self):
        pi=3.14
        return f'{pi*(self.r)**2*self.h:.2f}'
r,h=map(int,input().split())
x=Cylinder(r,h)
print(x.S(),x.V())
```
