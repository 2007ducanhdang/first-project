```python
#W7A16
import sys
from collections import Counter

def solve():
    input_data = sys.stdin.read().splitlines()
    for line in input_data:
        if not line.strip():
            continue
        arr = list(map(int, line.split()))
        counts = Counter(arr)
        remove_count = 0
        
        for x, freq in counts.items():
            if freq >= x:
                remove_count += freq - x
            else:
                remove_count += freq
                
        if remove_count == 0:
            print("YES")
        else:
            print(f"NO {remove_count}")

if __name__ == "__main__":
    solve()
```

```python
#W7A18
import sys

def solve():
    input_data = sys.stdin.read().splitlines()
    for line in input_data:
        if not line.strip():
            continue
        arr = list(map(int, line.split()))
        
        evens = sorted([x for x in arr if x % 2 == 0])
        odds = sorted([x for x in arr if x % 2 != 0], reverse=True)
        
        print(*(evens + odds))

if __name__ == "__main__":
    solve()
```

```python
#W7A19
import sys

def merge_arrays(a, b):
    ascending = True
    if len(a) >= 2:
        if a[0] > a[-1]: ascending = False
    elif len(b) >= 2:
        if b[0] > b[-1]: ascending = False
        
    i, j = 0, 0
    res = []
    
    while i < len(a) and j < len(b):
        # Nếu đang tăng dần
        if ascending:
            if a[i] < b[j]:
                res.append(a[i])
                i += 1
            else:
                res.append(b[j])
                j += 1
        else:
            if a[i] > b[j]:
                res.append(a[i])
                i += 1
            else:
                res.append(b[j])
                j += 1
                
    res.extend(a[i:])
    res.extend(b[j:])
    return res

def solve():
    lines = sys.stdin.read().splitlines()
    lines = [line.strip() for line in lines if line.strip()]
    for i in range(0, len(lines), 2):
        if i + 1 >= len(lines): break
        arr1 = list(map(int, lines[i].split()))
        arr2 = list(map(int, lines[i+1].split()))
        
        result = merge_arrays(arr1, arr2)
        print(*(result))

if __name__ == "__main__":
    solve()
```

```python
#W7A20
import sys

def solve():
    lines = sys.stdin.read().splitlines()
    lines = [line.strip() for line in lines if line.strip()]
    for i in range(0, len(lines), 2):
        if i + 1 >= len(lines): break
        
        arr_str = lines[i].split()
        try:
            k = int(lines[i+1])
        except ValueError:
            continue
            
        total_sum = 0
        for s in arr_str:
            if len(s) == k:
                total_sum += int(s)
                
        print(total_sum)

if __name__ == "__main__":
    solve()
```

```python
#W7A21
import sys

def solve():
    days_in_month = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    input_data = sys.stdin.read().split()
    
    iterator = iter(input_data)
    try:
        while True:
            day = int(next(iterator))
            month = int(next(iterator))
            
            if month < 1 or month > 12:
                print("invalid")
            elif day < 1 or day > days_in_month[month]:
                print("invalid")
            else:
                day_of_year = sum(days_in_month[:month]) + day
                print(day_of_year)
    except StopIteration:
        pass

if __name__ == "__main__":
    solve()
```

```python
#W7A22
import sys
from collections import Counter

def solve():
    lines = sys.stdin.read().splitlines()
    lines = [line.strip() for line in lines if line.strip()]
    idx = 0
    
    while idx < len(lines):
        try:
            n = int(lines[idx])
            idx += 1
        except ValueError:
            idx += 1
            continue
            
        names = []
        for _ in range(n):
            if idx < len(lines):
                full_name = lines[idx]
                words = full_name.split()
                if words:
                    names.append(words[-1])
                idx += 1
        
        if names:
            counts = Counter(names)
            sorted_names = sorted(counts.keys(), key=lambda x: (-counts[x], x))
            print(sorted_names[0])

if __name__ == "__main__":
    solve()
```

```python
#W7A24
import sys

def solve():
    lines = sys.stdin.read().splitlines()
    for line in lines:
        if not line.strip(): continue
        words = line.split()
        if not words:
            print("")
            continue
            
        res = [words[0]]
        for i in range(1, len(words)):
            if words[i] != words[i-1]:
                res.append(words[i])
        print(" ".join(res))

if __name__ == "__main__":
    solve()
```

```python
#W7A25
import sys

def solve():
    lines = sys.stdin.read().splitlines()
    lines = [line.strip() for line in lines if line.strip()]
    
    idx = 0
    while idx < len(lines):
        num_str = lines[idx]
        if idx + 1 >= len(lines): break
        try:
            k = int(lines[idx+1])
        except ValueError:
            break
            
        idx += 2
        

        stack = []
        to_remove = len(num_str) - k
        
        for digit in num_str:
            while to_remove > 0 and stack and stack[-1] < digit:
                stack.pop()
                to_remove -= 1
            stack.append(digit)
            
        while to_remove > 0:
            stack.pop()
            to_remove -= 1
            
        print("".join(stack))

if __name__ == "__main__":
    solve()
```

```python
#W7A26
import sys
import re

def solve():
    lines = sys.stdin.read().splitlines()
    for line in lines:
        if not line.strip(): continue
        
        numbers = re.findall(r'\d+', line)
        numbers = [int(x) for x in numbers]
        
        if numbers:
            total = sum(numbers)
            avg = total / len(numbers)
            if avg.is_integer():
                print(f"{total} {int(avg)}")
            else:
                print(f"{total} {avg}")

if __name__ == "__main__":
    solve()
```
