```python
def reverse (path: str) -> int:
    with open(path,'r') as f:
        c=f.read()
        reverse=c[::-1]
        return reverse
file='w1a11.txt'
with open(file,'w') as f:
    f.write(input())
kq=reverse(file)
print(kq)
    
```

```python
def canwinnim(path):
    with open(path,'r') as f:
        n=f.read()
        if int(n)%4==0:
            return False
        else:
            return True
file='w1a12.txt'
with open(file,'w') as f:
    f.write(input())
kq=canwinnim(file)
print(kq)
```

```python
def houseofhogwarts(path):
    with open(path,'r') as f:
        content=f.readlines()
        for i in content[1:]:
            print(i.strip())
file='w11a3.txt'
with open(file,'w') as f:
    n=input()
    f.write(n + '\n')
    for j in range(int(n)):                
        f.write(input()+'\n')
kq=houseofhogwarts(file)
```

```python
def areanagram(path):
    with open(path,'r') as f:
        content=f.readlines()
        data=[]
        for i in range(len(content)):
            data.append(sorted(list(content[i].strip().replace(' ',''))))
        if data[0]==data[1]:
            print(True)
        else:
            print(False)
        
        
file='w11a4.txt'
A='tom marvolo riddle'
B='i am lord voldemort'
with open(file,'w') as f:
    f.write(A+'\n')
    f.write(B+'\n')
kq=areanagram(file)
```

```python
def lettercount(path,n):
    with open(path,'r') as f:
        content=f.read()
        print(content.count(n))
file='w11a5.txt'
A='dsf sf'
a=input()
with open(file,'w') as f:
    f.write(A+'\n')
kq=lettercount(file,a)
```

```python
def fib(n):
    if n==0:
        return 'a'
    elif n==1:
        return 'b'
    else:
        return fib(n-1)+fib(n-2)
n=int(input())
for i in range(n+1):
    print(f'f({i})={fib(i)}')
```
